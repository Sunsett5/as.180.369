## **The Model**

### 1. Firms' revenue -> Reinvest in R&D

### 2. Total factor productivity may increase.

### 3. Lower cost -> Competitive pricing -> Gain market share.

### 4. Unprofitable firms exit the market. New firms enter.

### 5. Repeat until dynamic equilibrium.

## **Results - Overview**

### <ins>Set up</ins>

- 1 industry
- 10 Countries
- 20 Firms per country
- Lowest profitable 50% of all firms exit the industry each cycle.

## **Conclusion**

- ### Firms' decisions are heterogeneous

- ### Routines depend on R&D effectiveness

- ### Equilibrium is path-dependent

- ### Convergence is strongest with effective imitation

- ### Ineffective imitation leads to divergence and concentrated market
